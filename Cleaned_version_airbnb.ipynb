```python
# I begin by importing the necessary libraries
import pandas as pd
import numpy as np
from google.colab import files
```

```python
# Uploading 'airbnb.xlsx' dataset from folders into Google colab
file_path = files.upload()
```

```python
# Reading the excel file in pandas
df = pd.read_excel('airbnb.xlsx')
```

```python
# Display the dataset to confirm if it is uploaded
df.head(10)
```

```python
# Executing Exploratory Data Analysis(EDA) functions to give insights and errors about the data
df.info()
df.describe()
```

```python
# Executing Exploratory Data Analysis(EDA) functions to give missing values on the data
df.isnull().sum()
```

```python
# After proper EDA I first deal with missing values
df = df.dropna(subset=['Property Type'])
df = df.dropna(subset = ['Host Since'])
df['Review Scores Rating (bin)'] = df['Review Scores Rating (bin)'].fillna(df['Review Scores Rating (bin)'].mode()[0])
df['Beds'] = df['Beds'].fillna(df['Beds'].mode()[0])
df['Review Scores Rating'] = df['Review Scores Rating'].fillna(df['Review Scores Rating'].mean())
```

```python
# And I assign the correct data type to columns
df['Review Scores Rating (bin)'] = df['Review Scores Rating (bin)'].astype(np.int8)
df['Zipcode'] = df['Zipcode'].astype(str)
df['Beds'] = df['Beds'].astype(int)
df['Price'] = df['Price'].astype(float)
```

```python
# I then display my cleaned-version dataset
df.head(10)
```

```python
# Then I convert the 'df' dataset to CSV
df.to_csv('Cleaned_airbnb.csv', index = False,)
```
